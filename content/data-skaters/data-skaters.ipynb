# Data Skaters - Data Science

## Importing Our Data

```python
import piplite, pyodide
await piplite.install(['pandas', 'plotly', 'nbformat', 'statsmodels'])
import requests
import pandas as pd
import plotly.express as px

demographics_url = 'https://api.datadunkers.ca/api/collections/data_skaters_demographics/records'
response = requests.get(demographics_url, params={'perPage':500, 'page':1})
df_demographics = pd.DataFrame(response.json()['items'])
display(df_demographics)

url = 'https://api.datadunkers.ca/api/collections/data_skaters_activities/records'
first_page = requests.get(url, params={'perPage':500, 'page':1}).json()
item_list = first_page['items']
for n in range(2, first_page['totalPages'] + 1):
  item_list.extend(requests.get(url, params={'perPage':500, 'page':n}).json()['items'])
df_activities = pd.DataFrame(item_list)
display(df_activities)
```

### Filter for Today

```python
today = '2026-04-06'

df_demographics = df_demographics[df_demographics['created'].str.contains(today)]
df_activities = df_activities[df_activities['created'].str.contains(today)]
```

### Visualize

```python
px.scatter(df_demographics, x='resting_heart_rate', y='reaction_time_ms', trendline='ols', title='Reaction Time vs Resting Heart Rate')
```

```python
group_success_rate = df_activities.groupby('group_number')['success'].mean().reset_index()
px.bar(group_success_rate, x='group_number', y='success', title='Group Success Rate')
```

```python
px.imshow(df_demographics.corr(numeric_only=True), title='Correlation Matrix', height=800, text_auto='.2f')
```

## NHL Data

Let's look at data from the Winnipeg Jets 2025-2026 season.

```python
url = 'https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NHL/team_players/2025-2026/WPG_2025-2026_players.csv'
df_jets = pd.read_csv(pyodide.http.open_url(url))
df_jets
```

Now we can find the top ten players, by points, on the team.

```python
df_sorted = df_jets.sort_values(by='PTS', ascending=False)
df_sorted.head(10)
```

Then we can use Plotly Express to create a bar chart of the top ten players and their points.

```python
px.bar(df_sorted.head(10), x='Name', y='PTS',  title='Winnipeg Jets Top 10 Point Leaders (2025-2026)', labels={'Name': 'Player Name', 'PTS': 'Points'})
```

Next let's look at a larger dataset of all player statistics from the 2025-2026 season.

```python
url = 'https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NHL/player/nhl_player_stats_2025-2026.csv'
df_nhl = pd.read_csv(pyodide.http.open_url(url))
df_nhl
```

Then we can create a scatterplot of goals versus assists.

```python
px.scatter(df_nhl, x='G', y='A', hover_name='Name', title='NHL Goals vs Assists (2025-2026)', labels={'G': 'Goals', 'A': 'Assists'})
```

We can also filter the data set to just show the Winnipeg Jets.

```python
df_jets = df_nhl[df_nhl['Team'] == 'WPG']
df_jets
```

And then create a scatteplot of just the Jets data.

```python
px.scatter(df_jets, x='G', y='A', hover_name='Name', title='Jets Goals vs Assists (2025-2026)', labels={'G': 'Goals', 'A': 'Assists'})
```
